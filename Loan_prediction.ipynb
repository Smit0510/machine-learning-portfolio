```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
s_LP= StandardScaler()
```

```python
df=pd.read_csv('Loan Data.csv')
df
```

```python
df.info()
```

```python
df['Dependents'] = df['Dependents'].fillna('0')
df['Dependents'].unique()
```

```python
df['Gender'] = df['Gender'].fillna('Others')
df['Gender'].unique()
```

```python
df['Married'] = df['Married'].fillna('No')
df['Married'].unique()
```

```python
df['Self_Employed'] = df['Self_Employed'].fillna('No')
df['Self_Employed'].unique()
```

```python
df['LoanAmount'] = df['LoanAmount'].fillna('0').astype(int)
df['LoanAmount'].unique()
```

```python
df['Loan_Amount_Term'] = df['Loan_Amount_Term'].fillna('0').astype(int)
df['Loan_Amount_Term'].unique()
```

```python
df['Credit_History'] = df['Credit_History'].fillna('0').astype(int)
df['Credit_History'].unique()
```

```python
df.info()
```

```python
encoder = LabelEncoder()
df['Gender_E'] = encoder.fit_transform(df['Gender'])
df['Married_E'] = encoder.fit_transform(df['Married'])
df['Dependents_E'] = encoder.fit_transform(df['Dependents'])
df['Education_E'] = encoder.fit_transform(df['Education'])
df['Self_Employed_E'] = encoder.fit_transform(df['Self_Employed'])
df['Property_Area_E'] = encoder.fit_transform(df['Property_Area'])
df['Loan_Status_E'] = encoder.fit_transform(df['Loan_Status'])
df.head()
```

```python
x=df[['Gender_E','Married_E','Dependents_E','Education_E','Self_Employed_E','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History']]
y=df['Loan_Status_E']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25, random_state=43)
x_train.shape,x_test.shape
```

```python
x_train = s_LP.fit_transform(x_train)
x_test = s_LP.transform(x_test)
```

```python
knn_LP = KNeighborsClassifier(n_neighbors=8)
knn_LP.fit(x_train,y_train)
```

```python
y_pred = knn_LP.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
