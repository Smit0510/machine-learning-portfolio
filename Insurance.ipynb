```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
s=StandardScaler()
r=LinearRegression()
```

```python
df=pd.read_csv("insurance.csv")
df.head()
```

```python
df.info()
```

```python
cat_col=df.select_dtypes(include="object").columns
cat_col
```

```python
encode=LabelEncoder()
df['sexEncode']=encode.fit_transform(df['sex'])
df['smokerEncode']=encode.fit_transform(df['smoker'])
df['regionEncode']=encode.fit_transform(df['region'])
df
```

```python
plt.scatter(df['age'],df['charges'])
plt.xlabel('Age')
plt.ylabel('charges')
```

```python
x=df[['age']]
y=df['charges']
```

```python
x.shape,y.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.40,random_state=21)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train=s.fit_transform(x_train)
```

```python
x_test=s.fit(x_test)
```

```python
plt.scatter(x_train,y_train)
```

```python
r.fit(x_train,y_train)
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,r.predict(x_train), color="green")
plt.xlabel('age')
plt.ylabel('charges')
```
